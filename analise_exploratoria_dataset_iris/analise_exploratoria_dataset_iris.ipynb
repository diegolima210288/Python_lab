## Análise Exploratória no Dataset Iris

```python
# importando bibliotecas e pacotes
import time
import numpy as np
import pandas as pd
import matplotlib as mat
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline

fontsize = 14
ticklabelsize = 14
```

```python
# Carregando o dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns = iris.feature_names)
print(len(df))
df.head()
```

## Extração e Transformação de Dados

```python
# valores descritivos da Variável target
print(iris.target_names)
```

```python
# valores numéricos da Variável target
print(iris.target)
```

```python
# Adicionando ao dataset uma nova coluna com os nomes das espécies
ds_target = []
for i in iris.target:
    if i == 0:
        ds_target.append('setosa')
    elif i == 1:
        ds_target.append('versicolor')
    elif i == 2:
        ds_target.append('virginica')
        
df['ds_target'] = ds_target
```

```python
# Incluindo no dataset uma coluna com os valores numéricos da variável target
df['target'] = iris.target
```

```python
df.head()
```

```python
# Calcule a média de cada feature para as 3 classes
df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','ds_target']].groupby('ds_target').mean().T
```

## Plot

```python
# Histograma da features "sepal length (cm)"
df['sepal length (cm)'].hist();
```

```python
# Gráficos de Dispersão (scatter Plot) da variável sepal length versus número da linha
plt.scatter(data = df, x = range(len(df)), y = 'sepal length (cm)', c = df['target'])
plt.xlabel('Número de linhas')
plt.ylabel('sepal length (cm)')
plt.title('Gráfico Scatter Plot');
```

```python
# Gráficos de Dispersão (scatter Plot) da variável "sepal length (cm)" versus "sepal width (cm)"
plt.scatter(data = df, x = 'sepal length (cm)', y = 'sepal width (cm)', c = df['target'])
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Gráfico Scatter Plot');
```
